```python
#remove outliers

def outlier_limit(col):
 Q3,Q1=np.nanpercentile (col,[75,25])
 IQR=Q3-Q1
 UL=Q3+1.5*IQR
 LL=Q1-1.5*IQR
 return UL,LL
```

```python
for column in df.columns:
 if df[column].dtype!='object':
    UL,LL=outlier_limit(df[column])
    df[column]=np.where((df[column]>UL)|(df[column]<LL),np.nan,df[column])
```

```python
for i in df.columns:
 if df[i].dtypes != "object":
    sns.boxplot(y= df[i],color ='yellow')
    plt.show()
```

```python
#encoding categorical variables
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['Make'] = le.fit_transform(df['Make'])
df['Model'] = le.fit_transform(df['Model'])
df['Vehicle Class'] = le.fit_transform(df['Vehicle Class'])
df['Fuel Type'] = le.fit_transform(df['Fuel Type'])
df['Transmission'] = le.fit_transform(df['Transmission'])
df
```

```python
#Setting target and features
y = df[['CO2 Emissions(g/km)']].values
x=df.drop(['CO2 Emissions(g/km)'],axis=1).values
```

```python
#split the data for training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
#Normalization
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
x_test
```

```python
y_test
```

Model Creation

```python
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
y_pred1=lr_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_absolute_percentage_error
```

```python
score1=r2_score(y_test,y_pred1)
print("r2 score of Linear Regression Model:",score1)
```

```python
print('mean absolute error of LR model',mean_absolute_error(y_test,y_pred1))
print('mean absolute percentage error of LR model',mean_absolute_percentage_error(y_test,y_pred1))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model=RandomForestRegressor(n_estimators=100)
rf_model.fit(x_train,y_train)
y_pred2=rf_model.predict(x_test)
```

```python
score2=r2_score(y_test,y_pred2)
print("r2 score using Random Forest Model:",score2)
```

```python
print('mean absolute error of RF model',mean_absolute_error(y_test,y_pred2))
print('mean absolute percentage error of RF model',mean_absolute_percentage_error(y_test,y_pred2))
```

```python
from sklearn.tree import DecisionTreeRegressor
df_model=DecisionTreeRegressor()
df_model.fit(x_train,y_train)
y_pred3=df_model.predict(x_test)
```

```python
score3=r2_score(y_test,y_pred3)
print("r2 score of Decison Tree Model:",score3)
```

```python
print('mean absolute error of DT model',mean_absolute_error(y_test,y_pred3))
print('mean absolute percentage error of DT model',mean_absolute_percentage_error(y_test,y_pred3))
```

```python
New=pd.DataFrame({'Models':['LT','RF','DT'],'r2_score':[score1*100,score2*100,score3*100]})
New
```
